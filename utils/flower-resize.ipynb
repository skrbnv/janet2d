```python
import cv2
import numpy as np
import os
from glob import glob

src_path = '/media/sergey/faq/flower_data/valid'
tgt_path = '/media/sergey/faq/flower_data_resized/valid'
```

```python
subdirs = os.listdir(src_path)
for sd in subdirs:
    print(sd)
    if os.path.isdir(subdir := os.path.join(src_path, sd)):
        try:
            os.mkdir(os.path.join(tgt_path, sd))
        except:
            pass
        for jpeg in (jpegs := glob(os.path.join(subdir, '*.jpg'))):
            img = cv2.imread(jpeg)
            h, w, _ = img.shape
            ratio = min(h, w)/224
            nh, nw = int(h/ratio), int(w/ratio)
            img = cv2.resize(img, (nw, nh), interpolation = cv2.INTER_NEAREST)
            crop = img[(nh-224)//2:224+(nh-224)//2, (nw-224)//2:224+(nw-224)//2]
            assert crop.shape == (224,224,3), f'Wrong shape after resize/crop'
            cv2.imwrite(os.path.join(tgt_path,sd,os.path.basename(jpeg)), crop)
            print('#', end='')
    print('')

    
```
